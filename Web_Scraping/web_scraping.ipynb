```python
import csv
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq

my_url = "https://www.flipkart.com/search?q=xiaomi+mobile&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_0_2_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_0_2_na_na_na&as-pos=0&as-type=HISTORY&suggestionId=xiaomi+mobile%7Cin+Mobiles&requestId=768325c5-79c5-4c02-8f42-7d0890f2f146&as-backfill=on"

uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html,"html.parser")

name_of_product = page_soup.findAll("div",{"class":"_3wU53n"})
model = []
for i in range(0,len(name_of_product)):
     model.append(str(name_of_product[i]).split(">")[1].split("<")[0])

        
price_of_product = page_soup.findAll("div",{"class":"_1vC4OE _2rQ-NK"})
price = []
for i in range(0,len(price_of_product)):
    price.append(str(price_of_product[i]).split(">")[1].split("<")[0])

    
rating_of_product =  page_soup.findAll("div",{"class":"hGSR34"})
rating = []
for i in range(0,len(rating_of_product)):
    rating.append(str(rating_of_product[i]).split(">")[1].split("<")[0])


specification_of_product = page_soup.findAll("div",{"class":"_3ULzGw"})
specification = []
for k in range(0,len(specification_of_product)):
    b = specification_of_product[k].findAll("ul",{"class":"vFw0gD"})
    for i in range(0,len(b)):
        c = b[i].findAll("li",{"class":"tVe95H"})
        for j in range(0,len(c)):
            specification.append(str(c[j]).split(">")[1].split("<")[0])

            
import pandas as pd
df = pd.DataFrame()
df["Models"] = model
df["Price"] = price

df.to_csv("flipkart.csv")
```
